```python
def clean_data(data):
    row_to_drop = []
    for key,df in data.items():
        #standardize row index and column index
        if df.columns[0] == "代码":
            df.set_index('代码',inplace = True)
        elif df.columns[0] == "证券代码":
            df.set_index('证券代码',inplace = True)
        if key == "close" or key =="market_cap" or key=="PB_ratio":
            df.columns = pd.to_datetime(df.columns)
        
        #handling NaN problems
        if key!="index":
            df.replace(0, np.nan,inplace = True)
            for i in range(len(df)):
                if df.iloc[i,:].isna().sum().sum()>((len(df.columns)*1)/4):
                    if (i in row_to_drop)==False:
                        row_to_drop.append(df.index[i])
    #drop NaN rows
    for key, df in data.items():
        if key!="index":
            df.drop(row_to_drop,inplace=True)
            df.fillna(method="ffill",axis = 1,inplace= True)
            df.fillna(method="bfill",axis = 1,inplace = True)
    #handle total asset
    for key,df in data.items():
        if key == "total_asset":
            data[key] = data[key].pct_change(axis = "columns")
    
    #convert quarter data to daily data
    the_index = []
    the_columns = []
    for key,df in data.items():
        if key == "close":
            the_columns = df.columns
            the_index = df.index
        if key == "roe" or key =="total_asset" or key=="ESG_e" or key == "ESG_s" or key == "ESG_g":
            new_df = pd.DataFrame(np.nan,columns=the_columns,index = the_index)
            for i in df.columns:
                if i in new_df.columns:
                    new_df[i]= df[i]
                else:
                    temp_date = i
                    while (temp_date in new_df.columns) == False:
                        temp_date-=pd.Timedelta("1 days")
                    new_df[temp_date] = df[i]
            new_df.fillna(method="ffill",axis = 1,inplace= True)
            new_df.fillna(method="bfill",axis = 1,inplace = True)
            data[key] = new_df
    #handle bp 
    for key,df in data.items():
        if key == "PB_ratio":
            data[key] = 1/data[key]
    #transpose
    for key,df in data.items():
        if key!= 'index':
            data[key] = df.T
    #chnage to monthly data
    for key,df in data.items():
        if key != 'index':
            data[key]=df.resample('M').ffill()
        else:
            data[key].set_index('日期',inplace = True)
            data[key]=data[key].resample('M').ffill()
            temp = data[key].pct_change()
            data[key]=temp.shift(periods=-1)
            
    #timeline (starts from 2018-02-01)
    start_date = datetime.datetime.strptime('2018-02-28','%Y-%m-%d')
    for key,df in data.items():
        df_temp = df.loc[start_date:,:]
        data[key] = df_temp

        
    return data
def close_monthly_return (data):
    #calculate monthly return
    temp = data['close'].pct_change() 
    data['close_monthly_return'] = temp.shift(periods = -1)
    #drop the last row
    for key,df in data.items():
        df.drop(df.tail(1).index,inplace=True)
    #now every row of close_monthly_return has the return of the next month and every row of other df has the value of that month(index will be next month's return)
    return data
#divide stocks into 5*5 portfolios using size and d2 dimension
def divide_5_5 (month_return,size,d2):
    the_column = size.columns
    the_index = size.index
    size_table =  size.apply(pd.qcut, axis=1, args=([0, 0.2,0.4,0.6,0.8, 1.],),
                          labels=[0,1,2,3,4])
    d2_table = d2.apply(pd.qcut, axis=1, args=([0, 0.2,0.4,0.6,0.8, 1.],),
                          labels=[0,1,2,3,4])
    label_table = pd.DataFrame(columns = the_index,index =the_index)
    for i in the_index:
        for j in the_column:
            label_table.loc[i,j] = str(size_table.loc[i,j])+'*'+str(d2_table.loc[i,j])
    
    ans = {}
    company_book = {}
    for i in range(5):
        for j in range(5):
            key = str(i)+'*'+str(j)
            ans[key] = pd.DataFrame(columns = ['y'],index =the_index)
            company_book[key] = pd.DataFrame(columns = [key],index =the_index)
    for i in the_index:
        weight = size.loc[i,:].groupby(by=label_table.loc[i,:]).apply(list)
        for j in weight.index:
            weight.loc[j]=np.array(weight.loc[j])/sum(weight.loc[j])
        m_return = month_return.loc[i,:].groupby(by=label_table.loc[i,:]).apply(list)
        for j in m_return.index:
            m_return.loc[j]=np.array(m_return.loc[j])
            ans[j].loc[i,'y'] = weight.loc[j].dot(m_return.loc[j])
        company_name = the_column.to_series().groupby(by=label_table.loc[i,:]).apply(list)
        for j in company_name.index:
            company_name.loc[j] = np.array(company_name.loc[j])
            company_book[j].loc[i,j] = company_name.loc[j]
    division_table = pd.DataFrame( index =[0,1,2,3,4],columns= [0,1,2,3,4])
    for key,value in ans.items():
        indexs = key.split('*')
        for i in range(len(indexs)):
            indexs[i] =int(indexs[i])
        division_table.loc[indexs[0],indexs[1]] = ans[key]['y'].mean()
    return division_table, ans,company_book
#divide stocks into 2*3 porfolios using size and d2 dimension
def divide_2_3 (month_return,size,d2):
    the_column = size.columns
    the_index = size.index
    size_table =  size.apply(pd.qcut, axis=1, args=([0,0.5, 1.],),
                          labels=[0,1])
    d2_table = d2.apply(pd.qcut, axis=1, args=([0, 0.3,0.7,1.],),
                          labels=[0,1,2])
    label_table = pd.DataFrame(columns = the_index,index =the_index)
    for i in the_index:
        for j in the_column:
            label_table.loc[i,j] = str(size_table.loc[i,j])+'*'+str(d2_table.loc[i,j])
    
    ans = {}
    company_book = {}
    for i in range(2):
        for j in range(3):
            key = str(i)+'*'+str(j)
            ans[key] = pd.DataFrame(columns = ['y'],index =the_index)
            company_book[key] = pd.DataFrame(columns = [key],index =the_index)
    for i in the_index:
        weight = size.loc[i,:].groupby(by=label_table.loc[i,:]).apply(list)
        for j in weight.index:
            weight.loc[j]=np.array(weight.loc[j])/sum(weight.loc[j])
        m_return = month_return.loc[i,:].groupby(by=label_table.loc[i,:]).apply(list)
        for j in m_return.index:
            m_return.loc[j]=np.array(m_return.loc[j])
            ans[j].loc[i,'y'] = weight.loc[j].dot(m_return.loc[j])
        company_name = the_column.to_series().groupby(by=label_table.loc[i,:]).apply(list)
        for j in company_name.index:
            company_name.loc[j] = np.array(company_name.loc[j])
            company_book[j].loc[i,j] = company_name.loc[j]
    division_table = pd.DataFrame( index =[0,1],columns= [0,1,2])
    for key,value in ans.items():
        indexs = key.split('*')
        for i in range(len(indexs)):
            indexs[i] =int(indexs[i])
        division_table.loc[indexs[0],indexs[1]] = ans[key]['y'].mean()
    return division_table, ans,company_book
#generate factor
def factor_generator(data):
    size_value_table,size_value_dict,size_value_company = divide_2_3 (data['close_monthly_return'],data['market_cap'],data['PB_ratio'])
    SMB_value = pd.DataFrame()
    SMB_value['y'] = ((size_value_dict['0*2']['y']+size_value_dict['0*1']['y']+size_value_dict['0*0']['y'])/3)-((size_value_dict['1*2']['y']+size_value_dict['1*1']['y']+size_value_dict['1*0']['y'])/3)
    HML =pd.DataFrame()
    HML['y'] = ((size_value_dict['0*2']['y']+size_value_dict['1*2']['y'])/2)-((size_value_dict['0*0']['y']+size_value_dict['1*0']['y'])/2)
    
    size_profit_table,size_profit_dict,size_profit_company = divide_2_3 (data['close_monthly_return'],data['market_cap'],data['roe'])
    SMB_profit = pd.DataFrame()
    SMB_profit['y'] = ((size_profit_dict['0*2']['y']+size_profit_dict['0*1']['y']+size_profit_dict['0*0']['y'])/3)-((size_profit_dict['1*2']['y']+size_profit_dict['1*1']['y']+size_profit_dict['1*0']['y'])/3)
    RMW = pd.DataFrame()
    RMW['y']=((size_profit_dict['0*2']['y']+size_profit_dict['1*2']['y'])/2)-((size_profit_dict['0*0']['y']+size_profit_dict['1*0']['y'])/2)
    
    size_investment_table,size_investment_dict,size_investment_company  = divide_2_3 (data['close_monthly_return'],data['market_cap'],data['total_asset'])
    SMB_investment = pd.DataFrame()
    SMB_investment['y'] = ((size_investment_dict['0*2']['y']+size_investment_dict['0*1']['y']+size_investment_dict['0*0']['y'])/3)-((size_investment_dict['1*2']['y']+size_investment_dict['1*1']['y']+size_investment_dict['1*0']['y'])/3)
    CMA = pd.DataFrame()
    CMA['y'] = ((size_investment_dict['0*0']['y']+size_investment_dict['1*0']['y'])/2)-((size_investment_dict['0*2']['y']+size_investment_dict['1*2']['y'])/2)
    
    #factor from ESG_e (top, mid, bot) Top Minus Bot(TMB)
    size_e_table,size_e_dict,size_e_company = divide_2_3 (data['close_monthly_return'],data['market_cap'],data['ESG_e'])
    SMB_e = pd.DataFrame()
    SMB_e['y'] = ((size_e_dict['0*2']['y']+size_e_dict['0*1']['y']+size_e_dict['0*0']['y'])/3)-((size_e_dict['1*2']['y']+size_e_dict['1*1']['y']+size_e_dict['1*0']['y'])/3)
    TMB = pd.DataFrame()
    TMB['y'] = ((size_e_dict['0*2']['y']+size_e_dict['1*2']['y'])/2)-((size_e_dict['0*0']['y']+size_e_dict['1*0']['y'])/2)
    
    #factor from ESG_s (good,fine,bad) Good Minus Bad(GMB)
    size_s_table,size_s_dict ,size_s_company= divide_2_3 (data['close_monthly_return'],data['market_cap'],data['ESG_s'])
    SMB_s = pd.DataFrame()
    SMB_s['y'] = ((size_s_dict['0*2']['y']+size_s_dict['0*1']['y']+size_s_dict['0*0']['y'])/3)-((size_s_dict['1*2']['y']+size_s_dict['1*1']['y']+size_s_dict['1*0']['y'])/3)
    GMB = pd.DataFrame()
    GMB['y'] = ((size_s_dict['0*2']['y']+size_s_dict['1*2']['y'])/2)-((size_s_dict['0*0']['y']+size_s_dict['1*0']['y'])/2)    
    
    #factor from ESG_g (large,meadian,little) Large Minus Little(LML)
    size_g_table,size_g_dict ,size_g_company= divide_2_3 (data['close_monthly_return'],data['market_cap'],data['ESG_g'])
    SMB_g = pd.DataFrame()
    SMB_g['y'] = ((size_g_dict['0*2']['y']+size_g_dict['0*1']['y']+size_g_dict['0*0']['y'])/3)-((size_g_dict['1*2']['y']+size_g_dict['1*1']['y']+size_g_dict['1*0']['y'])/3)
    LML = pd.DataFrame()
    LML['y'] = ((size_g_dict['0*2']['y']+size_g_dict['1*2']['y'])/2)-((size_g_dict['0*0']['y']+size_g_dict['1*0']['y'])/2)      
    
    #SMB factor
    SMB = pd.DataFrame()
    SMB['y'] = (SMB_value['y']+SMB_profit['y']+SMB_investment['y']+SMB_e['y']+SMB_s['y']+SMB_g['y'])/6
    
    #market factor 
    MKT = pd.DataFrame()
    MKT['y'] = data['index'].iloc[:,0]
    return MKT,SMB,HML,RMW,CMA,TMB,GMB,LML

#generate statistics of factor data
def factor_analyzer(factor,factor_name):
    stats_table = pd.DataFrame(index = ['Mean','Std Dev'],columns=[factor_name])
    mean = factor['y'].mean()
    stats_table.loc['Mean',factor_name] = mean
    sd = factor['y'].std()
    stats_table.loc['Std Dev',factor_name] = sd
    stats_table.loc["Sharpe Ratio", factor_name] = mean / sd * math.sqrt(12)
    return stats_table

#generate factor covariance matrix
def factor_correlation_matrix(factor_dict):
    temp_matrix = pd.DataFrame()
    for key,df in factor_dict.items():
        temp_matrix[key] = df['y']
    temp_matrix = temp_matrix.astype('float64')
    corr_matrix = temp_matrix.corr()
    return temp_matrix,corr_matrix


def company_small_group(company_dict,dimension,name_list):
    #change the first line
    key = str(dimension)+'*0'
    df = company_dict[key]
    company_list = []
    for i in df.index:
        company_list=company_list+df.loc[i,key].tolist()
    company_list_final =  np.unique(np.array(company_list)).tolist()
    return name_list.loc[company_list_final].values.tolist()


def company_high_group(company_dict,dimension,name_list):
    #change the first line
    key = str(dimension)+'*4'
    df = company_dict[key]
    company_list = []
    for i in df.index:
        company_list=company_list+df.loc[i,key].tolist()
    company_list_final =  np.unique(np.array(company_list)).tolist()
    return name_list.loc[company_list_final].values.tolist()

def regression_compare(regression_list,factor_matrix,y_dict):
#     factor_matrix = factor_matrix.shift(periods=1)
#     factor_matrix.drop(factor_matrix.index[0],inplace = True)
#     for key,df in y_dict.items():
#         y_dict[key] = df.astype('float64')
#         y_dict[key].drop(y_dict[key].index[0],inplace = True)

    for key,df in y_dict.items():
        df.fillna(method="ffill",inplace= True)
        y_dict[key] = df.astype('float64')     

    abs_r_list = [] 
    mkt = factor_matrix['MKT'].to_frame()
    for key,df in y_dict.items():
        reg = LinearRegression().fit(mkt, df)
        abs_r_list.append(abs(reg.intercept_))
    average_r =  (sum(abs_r_list) / len(abs_r_list))[0]
    result_table = pd.DataFrame(columns = ['abs_alpha_avg','abs_alpha_avg/avs_r_avg'],index = regression_list)
    for i in regression_list:
        feature_to_use = i.split(' ')
        X = factor_matrix[feature_to_use]
        alpha_list = []
        for key,df in y_dict.items():
            reg = LinearRegression().fit(X, df)
            alpha_list.append(abs(reg.intercept_))      
        alpha_avg = (sum(alpha_list) / len(alpha_list))[0]
        result_table.loc[i,'abs_alpha_avg'] = round(alpha_avg,4)
        result_table.loc[i,'abs_alpha_avg/avs_r_avg'] = str(round(alpha_avg/average_r *100,2))+'%'
    return result_table


def inter_regression (factor_matrix):
    result_table = pd.DataFrame(columns = ['intercept','MKT','SMB','HML','RMW','CMA','TMB','GMB','LML','R^2'],index = [
        'MKT','SMB','HML','RMW','CMA','TMB','GMB','LML'])
    length = len(factor_matrix.columns)
    for i in range(length):
        y = factor_matrix.iloc[:,i].to_frame()
        X = factor_matrix.iloc[:, np.r_[0:i, i+1:length]]
        reg = LinearRegression().fit(X, y)
        intercept = reg.intercept_       
        score = reg.score(X,y)
        coef=reg.coef_
        columns = result_table.columns
        count=0
        for j in range(len(columns)):
            if j == 0:
                result_table.iloc[i,j] = round(intercept[0],3)
            elif j == len(columns)-1:
                result_table.iloc[i,j] = round(score,3)
            elif j == i+1:
                result_table.iloc[i,j] = ' '
            else:
                result_table.iloc[i,j] =round(coef[0][count],3)
                count+=1
    return result_table





def regression_detail(factor_matrix,y_dict,required_coef):
    for key,df in y_dict.items():
        df.fillna(method="ffill",inplace= True)
        y_dict[key] = df.astype('float64')     

    result_table = pd.DataFrame( index =[0,1,2,3,4],columns= [0,1,2,3,4])
    
    for key,df in y_dict.items():
        indexs = key.split('*')
        for i in range(len(indexs)):
            indexs[i] =int(indexs[i])
        reg =  LinearRegression().fit(factor_matrix, df)
        alpha = round(reg.intercept_[0],3)
        coef = reg.coef_
        if required_coef == 'intercept':
            result_table.loc[indexs[0],indexs[1]] = alpha
        if required_coef == 'MKT':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][0],3)  
        if required_coef == 'SMB':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][1],3)   
        if required_coef == 'HML':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][2],3)  
        if required_coef == 'RMW':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][3],3)  
        if required_coef == 'CMA':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][4],3)   
        if required_coef == 'TMB':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][5],3)  
        if required_coef == 'GMB':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][6],3)  
        if required_coef == 'LML':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][7],3)  
    return result_table




def regression_detail_five_factor(factor_matrix,y_dict,required_coef):
    for key,df in y_dict.items():
        df.fillna(method="ffill",inplace= True)
        y_dict[key] = df.astype('float64')     

    result_table = pd.DataFrame( index =[0,1,2,3,4],columns= [0,1,2,3,4])
    
    for key,df in y_dict.items():
        indexs = key.split('*')
        for i in range(len(indexs)):
            indexs[i] =int(indexs[i])
        X = factor_matrix.iloc[:,0:5] #change here if model changes
        reg =  LinearRegression().fit(X, df)
        alpha = round(reg.intercept_[0],3)
        coef = reg.coef_
        if required_coef == 'intercept':
            result_table.loc[indexs[0],indexs[1]] = alpha
        if required_coef == 'MKT':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][0],3)  
        if required_coef == 'SMB':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][1],3)   
        if required_coef == 'HML':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][2],3)  
        if required_coef == 'RMW':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][3],3)  
        if required_coef == 'CMA':
            result_table.loc[indexs[0],indexs[1]] = round(coef[0][4],3) #add or del if model changes  
    return result_table


def generate_graph(factor_matrix):
    columns =factor_matrix.columns
    df = pd.DataFrame(columns = columns)
    for i in columns:
        df[i] = (1 + factor_matrix[i]).cumprod() - 1
    plt.figure();
    df.plot();
```

```python
def fix_and_print_latex(df):
    df.loc['Total'] = df.sum(axis=0)
    df.loc[:,'Total'] = df.sum(axis=1)
    df = df.round(4)
    df[0] = df[0].astype('double')
    df[1] = df[1].astype('double')
    df[2] = df[2].astype('double')
    df[3] = df[3].astype('double')
    df[4] = df[4].astype('double')
    df = df.round(4)
    return df.to_latex()
```
